```python
### Estructuras de datos
#### Lista

a = [1,2,3,4,False,'hola']
print(type(a))
print(a[0])
print(a[5])
print(a[-1])
```

```python
### Agregar valores

a = 1
b = [2,4,6,9]

b.append(a)

print(b)
```

```python
### Operaciones con sets

a = {1,1,1,1,1,1,2,2,5,5}
b = {5,10,15}

print(a | b) ## Union
print(a & b) ## Intersección
```

```python
### Diccionarios

a = {'entrada':'causa',
     'plato_fondo':['lomo saltado','arroz con pollo','mondonguito'],
     'postre':['mazamorra','arroz con leche'],
     'precio': 15.5,
     'descuento_estudiante': False}

print(a)
print(type(a))
print(a['precio'])
print(a['postre'])
print(a.keys())
print(a.values())
```

```python
### Estructuras de control

### Condicional
nota = 10
paneton = False
nombre = 'Johan'
if nota > 13:
  print('Aprobado')
elif paneton == True:
  print('Aprobado y recomendado para GEA')
elif nombre == 'Miluska':
  print('Aprobado en décimo superior')
else:
  print('Desaprobado')
```

```python
### Ciclos finitos

a = [2,4,6,8,10]

for valor in a:
  print(valor**2)
```

```python
a = 'hola'
for i in a:
  print(i)
```

```python
### Ciclo infinito

nota = 7

while nota < 13:
  print('Alumno debe ir a suplicatorio')
  print(nota)
  nota = nota + 1
```
