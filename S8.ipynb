```python
data = pd.read_csv('pelis_recommend.csv')
data.head()
```

```python
### Peliculas que se parezcan en el título
#!pip install unidecode
from sklearn.feature_extraction.text import CountVectorizer
import re
from unidecode import unidecode
from sklearn.metrics.pairwise import cosine_similarity
def limpiar_texto(texto):
  texto = texto.lower()
  texto = unidecode(texto)
  texto = re.sub(r'[^a-zA-Z\s]', '', texto)
  return texto
data['titulo_limpio'] = data['titulo'].apply(limpiar_texto)
vect = CountVectorizer()
a = pd.DataFrame(vect.fit_transform(data['titulo_limpio']).toarray(), columns = vect.get_feature_names_out())
matriz = pd.DataFrame(cosine_similarity(a,a), index = data['titulo'].values, columns = data['titulo'].values)
def recomendacion(titulo, n ):
  return matriz.loc[titulo,:].sort_values(ascending = False)[1:n+1]

recomendacion("The Time Traveler's Wife", 10)
```

```python
### Si estamos hablando de distancias, tenemos que escalar

from sklearn.preprocessing import StandardScaler

VARS = ['duration','popularity','danceability','acousticness','energy','instrumentalness','liveness','loudness','speechiness','Polarity']

esc = StandardScaler()

canciones_escalado = pd.DataFrame(esc.fit_transform(canciones[VARS]), columns = VARS)

matriz_escalado = pd.DataFrame(cosine_similarity(canciones_escalado, canciones_escalado), index = canciones['name'].values, columns =canciones['name'].values)

def recomendador_canciones(cancion, n):
  return matriz_escalado.loc[cancion,:].sort_values(ascending = False)[1:n+1]

recomendador_canciones('LALA',10)
```
